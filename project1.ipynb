```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read dataframe
fish = pd.read_csv('./data/Fish.csv')
```

```python
# Data sample
fish.head()
```

```python
# Data info
fish.info()
```

```python
# Data describe: Measures of dispersion
fish.describe()
```

```python
# View columns (header)
fish.columns
```

```python
# Rename header
fish.rename(columns={'Length1': 'Standard length', 'Length2': 'Fork length', 'Length3': 'Total length'})
```

```python
# Graphs relating to species and characteristics

# Species x Weight
sns.jointplot(data = fish , x = 'Species' , y = 'Weight' , hue = 'Species' , kind = 'scatter' , marginal_ticks=False)
                                                                                                                                    
plt.show()
```

```python
# Sum_total_for_specie
Sum_total_for_specie = fish.groupby('Species')['Weight'].sum()
print(Sum_total_for_specie)
```

```python
# Species x Total length
sns.pairplot(data = fish ,hue = 'Species', palette='coolwarm',corner=True, kind='hist')
plt.show()
```
